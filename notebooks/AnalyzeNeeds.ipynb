```python
# san_pietro_path_builder.py – build shortest paths from San Pietro houses (per-house progress bars)
# ---------------------------------------------------------------------------
# Assumes the following Geo/DataFrames are already in the notebook namespace:
#   1. san_pietro_casas   – GeoDataFrame of house origin points
#   2. bottega_std_groups – DataFrame mapping social-class needs → PP_Bottega_STD codes
#   3. needs              – GeoDataFrame of POI targets (workshops/services)
# Each POI *and* house row must carry the columns:  uid, row_adj, col_adj, geometry.
# src_cpp.path_planner.path_from_poi (C++ A*) must be importable.
# ---------------------------------------------------------------------------

from __future__ import annotations

import multiprocessing as mp
from concurrent.futures import ThreadPoolExecutor, as_completed
from pathlib import Path
import time
import logging

import numpy as np
import pandas as pd
import geopandas as gpd
from tqdm.notebook import tqdm  # Jupyter-friendly progress bar

# --------------------------- CONFIG ---------------------------------------
GRID_PATH  = "../cpp_data/grid.npy"          # uint8 walkable grid produced by rasteriser
OUTPUT_DIR = "../paths_npz_needs_santa_croce"           # .npz files + index CSV
INDEX_FILE = Path(OUTPUT_DIR) / "connection_index.csv"

RADIUS_M   = 1260                            # metres – search radius for candidate POIs
TOP_N      = 2                               # top-N nearest POIs *per* needed service
MAX_DIST   = -1                              # unlimited cut-off inside A*

VERBOSE    = True                            # flip to False for quieter output

# --------------------------- logging --------------------------------------
log_level = logging.DEBUG if VERBOSE else logging.INFO
logging.basicConfig(format="[%(levelname)s] %(message)s", level=log_level, force=True)
logger = logging.getLogger(__name__)

# --------------------------- cache helpers --------------------------------

def load_index() -> pd.DataFrame:
    if INDEX_FILE.exists():
        logger.info(f"Loading existing connection index → {INDEX_FILE}")
        return pd.read_csv(INDEX_FILE)
    logger.info("No existing index found – starting fresh.")
    return pd.DataFrame(columns=[
        "origin_uid", "origin_type",
        "target_uid", "target_type",
        "path_len",   "path_file",
    ])


def add_rows_to_index(df_index: pd.DataFrame, new_rows: list[dict]) -> pd.DataFrame:
    if not new_rows:
        return df_index
    df_index = pd.concat([df_index, pd.DataFrame(new_rows)], ignore_index=True)
    INDEX_FILE.parent.mkdir(parents=True, exist_ok=True)
    df_index.to_csv(INDEX_FILE, index=False)
    logger.debug(f"Indexed {len(new_rows)} new CASA→POI connections.")
    return df_index

# --------------------------- path helpers ---------------------------------
from src_cpp.path_planner import path_from_poi  # C++ A*

def compute_path(grid: np.ndarray, sr: int, sc: int, trg_r: int, trg_c: int):
    found = path_from_poi(grid, sr, sc, trg_r, trg_c,
                          max_distance=MAX_DIST, diagonals=True)
    if found is None:
        return None
    rows, cols = found
    return len(rows), rows, cols

# --------------------------- main -----------------------------------------

def main() -> None:
    t0 = time.time()

    # 1 | Ensure metre CRS ---------------------------------------------------
    for name, gdf in (("casas", san_pietro_casas), ("needs", needs)):
        if not gdf.crs or not gdf.crs.axis_info[0].unit_name.lower().startswith("metre"):
            logger.info(f"Converting CRS for {name} to metre-based …")
            gdf.to_crs(gdf.estimate_utm_crs(), inplace=True)

    # 2 | Spatial index ------------------------------------------------------
    needs_sindex = needs.sindex
    logger.info(f"Spatial index built for {len(needs):,} POIs.")

    # 3 | Load grid ----------------------------------------------------------
    GRID = np.load(GRID_PATH)
    logger.info(f"Grid loaded – shape {GRID.shape}")

    # 4 | Resume cache -------------------------------------------------------
    df_index  = load_index()
    done_uid  = set(df_index["origin_uid"].unique())

    # 5 | Thread pool --------------------------------------------------------
    n_threads = mp.cpu_count()
    executor  = ThreadPoolExecutor(max_workers=n_threads)

    # 6 | Build list of houses (unique uid) ----------------------------------
    casas_todo = (
        san_pietro_casas
          .loc[~san_pietro_casas["uid"].isin(done_uid)]
          .drop_duplicates(subset="uid", keep="first")
    )
    total_houses = len(casas_todo)
    if total_houses == 0:
        logger.warning("Nothing to do – all houses cached.")
        return

    outer = tqdm(total=total_houses, desc="Houses", position=0)

    paths_written, houses_skipped = 0, 0

    for _, casa in casas_todo.iterrows():
        uid = casa["uid"]
        outer.set_description_str(f"House {uid}")

        sr, sc = int(casa["row_adj"]), int(casa["col_adj"])

        # 6.1 | Needs --------------------------------------------------------
        needs_codes: list[str] = []
        if casa["is_nobility_ten"]:
            needs_codes += bottega_std_groups.loc[
                bottega_std_groups["is_nobility_need"], "PP_Bottega_STD"].tolist()
        if casa["is_middle_ten"]:
            needs_codes += bottega_std_groups.loc[
                bottega_std_groups["is_middle_class_need"], "PP_Bottega_STD"].tolist()
        if casa["is_lower_class_ten"]:
            needs_codes += bottega_std_groups.loc[
                bottega_std_groups["is_lower_class_need"], "PP_Bottega_STD"].tolist()
        if not needs_codes:
            houses_skipped += 1
            outer.update(); continue

        # 6.2 | Candidate POIs ----------------------------------------------
        buffer       = casa.geometry.buffer(RADIUS_M)
        cand_idx     = list(needs_sindex.intersection(buffer.bounds))
        cand_df      = needs.iloc[cand_idx].copy()
        cand_df      = cand_df[cand_df["PP_Bottega_STD"].isin(needs_codes)]
        cand_df["distance_m"] = cand_df.geometry.distance(casa.geometry)
        cand_df      = cand_df[cand_df["distance_m"] <= RADIUS_M]
        if cand_df.empty:
            houses_skipped += 1
            outer.update(); continue

        cand_df = (cand_df.sort_values("distance_m")
                           .groupby("PP_Bottega_STD", group_keys=False)
                           .head(TOP_N))

        # 6.3 | Paths --------------------------------------------------------
        fut_map = {
            executor.submit(compute_path, GRID, sr, sc,
                            int(row.row_adj), int(row.col_adj)): (row.PP_Bottega_STD, row.uid)
            for _, row in cand_df.iterrows()
        }
        inner = tqdm(total=len(fut_map), desc="services", position=1, leave=False)

        best: dict[str, dict] = {}
        for fut in as_completed(fut_map):
            inner.update()
            res = fut.result()
            if res is None:
                continue
            L, rows, cols = res
            code, tgt_uid = fut_map[fut]
            cur = best.get(code)
            if cur is None or L < cur["len"]:
                best[code] = dict(len=L, uid=tgt_uid, rows=rows, cols=cols)
        inner.close()

        # 6.4 | Persist ------------------------------------------------------
        if best:
            Path(OUTPUT_DIR).mkdir(exist_ok=True)
            new_rows = []
            for code, d in best.items():
                fname = f"{uid}_{d['uid']}.npz"
                fpath = Path(OUTPUT_DIR) / fname
                np.savez_compressed(fpath, rows=d["rows"], cols=d["cols"])
                # Remove any leading "../" or "./" from OUTPUT_DIR for the path_file string
                out_dir_clean = OUTPUT_DIR.lstrip("./")
                rel_path = f"{out_dir_clean}/{fname}"
                new_rows.append(dict(
                    origin_uid=uid, origin_type="CASA",
                    target_uid=d["uid"], target_type=code,
                    path_len=d["len"], path_file=rel_path))
            df_index = add_rows_to_index(df_index, new_rows)
            paths_written += len(new_rows)
        else:
            houses_skipped += 1

        outer.update()

    executor.shutdown(wait=True)
    outer.close()

    dt = time.time() - t0
    logger.info(
        f"✓ Finished {total_houses:,} houses in {dt/60:.1f} min – "
        f"{paths_written:,} paths saved, {houses_skipped:,} houses skipped.")


if __name__ == "__main__":
    main()
```
