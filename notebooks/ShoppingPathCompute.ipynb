```python
# Daily errands – perishables & daily use
daily = [
    "PISTOR",      # bread
    "GALINER",     # poultry
    "FRUTARIOL",   # fruit & veg
    "NARANZER",    # citrus seller
    "ERBARIOL",    # herbs/greens
    "POZZO",       # well
    "MALVASIA",    # wine shop
    "CAFFÈ",       # coffeehouse
    "ACQUAVITA",   # spirits
]

# Weekly errands – staples that keep a few days, grooming, small notions
weekly = [
    "LUGANEGHER",  # sausages
    "MARZER",      # butcher/cured meats
    "FORMAGIER",   # cheese
    "CASARIOL",    # dairy
    "LASAGNER",    # pasta maker
    "BIAVARIOL",   # grain merchant
    "SCALETER",    # pastry/sweets
    "MANDOLER",    # almond-confectioner
    "BARBIER",     # barber
    "LINARIOL",    # linen-seller
    "SARTOR",      # minor mends
    "GUA",         # haberdasher
    "SPECIER",     # everyday spices
    "TABACCO",     # tobacco
]

# Monthly / occasional errands – durable goods & major services
monthly = [
    "CALEGHER",    # fine shoemaker
    "ZAVATER",     # rough cobbler
    "TENTOR",      # dyer / washer
    "FABRO",       # blacksmith
    "CHIODI",      # nails / hardware
    "PIGNATER",    # earthenware
    "MASTELLER",   # tubs & barrels
    "STRAMAZER",   # mattress stuffer
    "MEDICINE",    # apothecary
]
```

```python
# ────────────────────────────────────────────────────────────────────────────
# NOTEBOOK CELL — build daily / weekly / monthly salesman tour
# ────────────────────────────────────────────────────────────────────────────
import numpy as np, pandas as pd
from pathlib import Path
from src.calculate_poi_paths import path_through_pois as _exact
from src_cpp.path_planner import path_from_poi           # C++ A*

# ---------------------------------------------------------------------------
# # 0) frequency tables (copy-paste yours here)
# daily   = ["PISTOR","GALINER","FRUTARIOL","NARANZER","ERBARIOL",
#            "POZZO","MALVASIA","CAFFÈ","ACQUAVITA"]
# weekly  = ["LUGANEGHER","MARZER","FORMAGIER","CASARIOL","LASAGNER","BIAVARIOL",
#            "SCALETER","MANDOLER","BARBIER","LINARIOL","SARTOR","GUA",
#            "SPECIER","TABACCO"]
# monthly = ["CALEGHER","ZAVATER","TENTOR","FABRO","CHIODI","PIGNATER",
#            "MASTELLER","STRAMAZER","MEDICINE"]
# ---------------------------------------------------------------------------

# 1) attach a `frequency` column
schedule = ({c:"daily" for c in daily} |
            {c:"weekly" for c in weekly} |
            {c:"monthly" for c in monthly})
shop_df  = current_house_shopping_list.copy()
shop_df["frequency"] = shop_df["target_type"].map(schedule).fillna("monthly")

# 2) choose which tour
freq_to_keep = "monthly"              # ← "weekly" / "monthly"
todo_df = shop_df.query("frequency == @freq_to_keep")

# 3) build target_POIs  (row,col of last step in each path file)
def last_cell(npz_rel_path):
    data = np.load(Path("..") / npz_rel_path, mmap_mode="r")
    return int(data["rows"][-1]), int(data["cols"][-1])

target_POIs = {
    row.target_type: last_cell(row.path_file)
    for _, row in todo_df.iterrows()
}
if not target_POIs:
    raise ValueError(f"No POIs tagged {freq_to_keep!r} for this house.")

poi_labels = list(target_POIs.keys())
poi_rcs    = list(target_POIs.values())

# 4) fast greedy solver (nearest-neighbour + return)  ─────────────────────
def greedy_tour(grid, start_rc, poi_rcs):
    sr, sc = start_rc
    remain = poi_rcs.copy()
    order  = []
    full_r, full_c = [], []
    while remain:
        # pick Euclidean-closest next POI (no path yet)
        idx = min(range(len(remain)),
                  key=lambda i: (remain[i][0]-sr)**2 + (remain[i][1]-sc)**2)
        trg_r, trg_c = remain.pop(idx)
        seg = path_from_poi(grid, sr, sc, trg_r, trg_c)
        if seg is None: return None     # disconnected
        rows, cols = seg
        full_r.extend(rows); full_c.extend(cols)
        order.append(poi_rcs.index((trg_r, trg_c)))
        sr, sc = trg_r, trg_c
    # return to start
    seg = path_from_poi(grid, sr, sc, *start_rc)
    if seg is None: return None
    rows, cols = seg
    full_r.extend(rows); full_c.extend(cols)
    return np.array(full_r,int), np.array(full_c,int), order

# 5) wrapper: try exact Held-Karp, fall back to greedy  ───────────────────
def path_through_pois_safe(grid, start_rc, poi_rcs):
    if not poi_rcs:
        return np.array([],int), np.array([],int), []
    try:
        return _exact(grid, start_rc, poi_rcs)
    except Exception as e:
        print("Exact solver failed ➜ using greedy:", e)
        out = greedy_tour(grid, start_rc, poi_rcs)
        if out is None:
            raise RuntimeError("Greedy solver also failed (disconnected graph)")
        return out

# 6) compute tour
start_rc = (int(current_house_poi.row_adj.iloc[0]),
            int(current_house_poi.col_adj.iloc[0]))

path_r, path_c, visit_order = path_through_pois_safe(grid, start_rc, poi_rcs)

print("visit order:", [poi_labels[i] for i in visit_order])
print("total steps:", len(path_r))

# 7) save for later viz
np.save("../case_study/salesman_rows.npy", path_r)
np.save("../case_study/salesman_cols.npy", path_c)
np.save("../case_study/visit_order.npy", visit_order)
```

```python
# ─────────────────────────────────────────────────────────────────────────────
# NOTEBOOK CELL — “high-res” sliding-window movie
#
#  • every GRID CELL is rendered as SCALE×SCALE *pixels*
#  • gold trail, yellow current pixel, peach halo, green POI dots
#  • on arrival at a POI the movie pauses 3 s and shows the POI name
#    centred on screen (Helvetica or fallback)
#
#  INPUTS that must already exist:
#      grid                  uint8  (H×W)
#      shopping_path_rows    1-D int, len N
#      shopping_path_cols    1-D int, len N
#      poi_rcs               list[(row,col)]
#      visit_order           list[int]   indices into poi_rcs
#      poi_labels  (optional) list[str]  same length as poi_rcs
# ─────────────────────────────────────────────────────────────────────────────
import numpy as np, shutil, os, imageio.v2 as imageio
from pathlib import Path
from PIL import Image, ImageDraw, ImageFont

# ─── parameters you may tweak ─────────────────────────────────────────────
SCALE        = 4                     # ⬅️  pixels per grid cell (resolution)
WIN_W, WIN_H = 720, 480              # window size in GRID CELLS
FPS          = 24
PAUSE_SEC    = 3
NEIGH_RAD    = 1
PATH_RGB     = (255, 215,  0)        # gold
CURR_RGB     = (255, 255,  0)        # yellow
HALO_RGB     = (255, 180,100)        # peach
DOT_RGB      = (  0, 255,  0)        # green
WORK_DIR     = Path("tmp_frames_npz")
VIDEO_OUT    = "tour_sliding_window.mp4"

# font (bold - large)
try:
    FONT = ImageFont.truetype("arialbd.ttf", 144)
except OSError:
    FONT = ImageFont.truetype("DejaVuSans-Bold.ttf", 144) \
           if os.path.exists("/usr/share/fonts/truetype/dejavu/DejaVuSans-Bold.ttf") \
           else ImageFont.load_default()

# font for subtitle (smaller)
try:
    FONT_SUB = ImageFont.truetype("arial.ttf", 64)
except OSError:
    FONT_SUB = ImageFont.truetype("DejaVuSans.ttf", 64) \
        if os.path.exists("/usr/share/fonts/truetype/dejavu/DejaVuSans.ttf") \
        else ImageFont.load_default()



# land-use palette → RGB LUT
PALETTE = {0:(173,216,230), 1:(190,190,190), 2:(178,34,34),
           3:( 64,224,208), 4:(152,251,152)}
lut = np.zeros((256,3), np.uint8)
for k,v in PALETTE.items(): lut[k]=v

# ─── derived values ───────────────────────────────────────────────────────
H, W          = grid.shape
rows, cols    = shopping_path_rows.astype(int), shopping_path_cols.astype(int)
N             = len(rows)
PAUSE_FRAMES  = int(PAUSE_SEC*FPS)
PIX_W, PIX_H  = WIN_W*SCALE, WIN_H*SCALE     # pixel dimensions of each frame

labels = poi_labels if "poi_labels" in globals() \
         else [f"POI {i}" for i in range(len(poi_rcs))]
label_by_coord = {tuple(poi_rcs[i]): labels[i] for i in range(len(poi_rcs))}

# fresh work dir
if WORK_DIR.exists(): shutil.rmtree(WORK_DIR)
WORK_DIR.mkdir()

overlay_trail = np.zeros((H, W, 3), np.uint8)   # cell-level
overlay_dots  = np.zeros((H, W, 3), np.uint8)
visited_labels = set()

def clamp(v, lo, hi): return max(lo,min(v,hi))

frame_idx=0
poi_ptr  =0

# ─────────────────────────── FIRST PASS ──────────────────────────────────
for r,c in zip(rows,cols):
    # 1 | window bounds (cell coords)
    r0 = clamp(r-WIN_H//2, 0, H-WIN_H)
    c0 = clamp(c-WIN_W//2, 0, W-WIN_W)
    r1,c1 = r0+WIN_H, c0+WIN_W
    lr,lc = r-r0, c-c0                    # local cell coords

    # 2 | compose CELL-level frame
    cell_frame = lut[grid[r0:r1, c0:c1]].copy()
    traj = overlay_trail[r0:r1, c0:c1];  mask=traj.any(-1)
    cell_frame[mask]=traj[mask]
    dots = overlay_dots[r0:r1, c0:c1];   dmask=dots.any(-1)
    cell_frame[dmask]=dots[dmask]

    # 2.1 | current pixel + halo on cell grid
    for nr in range(lr-NEIGH_RAD, lr+NEIGH_RAD+1):
        if not (0<=nr<WIN_H): continue
        for nc in range(lc-NEIGH_RAD, lc+NEIGH_RAD+1):
            if not (0<=nc<WIN_W): continue
            cell_frame[nr,nc]= CURR_RGB if (nr,nc)==(lr,lc) else HALO_RGB

    # 3 | upscale to PIXELS   (nearest-neighbor via repeat)
    frame = np.repeat(np.repeat(cell_frame, SCALE, 0), SCALE, 1)

    # 4 | save step frame
    np.savez_compressed(WORK_DIR/f"{frame_idx:06d}.npz", frame=frame)
    frame_idx+=1

    # 5 | arrival at POI ?
    if poi_ptr<len(visit_order) and (r,c)==tuple(poi_rcs[visit_order[poi_ptr]]):
        overlay_dots[r,c]=DOT_RGB
        msg = label_by_coord[(r,c)]
        visited_labels.add(msg)

        # Get English translation for subtitle
        subtitle = translation.get(str(msg), "")

        for _ in range(PAUSE_FRAMES):
            pil = Image.fromarray(frame.copy())   # PIXEL-level image
            draw=ImageDraw.Draw(pil)
            # centre label
            tb=draw.textbbox((0,0),msg,font=FONT)
            tw,th = tb[2]-tb[0], tb[3]-tb[1]
            cx,cy = PIX_W//2 - tw//2, PIX_H//2 - th//2
            draw.text((cx,cy), msg, font=FONT, fill=(255,255,255))
            # Draw subtitle (English) below the Italian label, in parenthesis
            if subtitle:
                sub_msg = f"({subtitle})"
                tb_sub = draw.textbbox((0,0), sub_msg, font=FONT_SUB)
                tw_sub, th_sub = tb_sub[2]-tb_sub[0], tb_sub[3]-tb_sub[1]
                cx_sub = PIX_W//2 - tw_sub//2
                cy_sub = cy + th + 20  # 20 px below main label
                draw.text((cx_sub, cy_sub), sub_msg, font=FONT_SUB, fill=(255,255,255))
            np.savez_compressed(WORK_DIR/f"{frame_idx:06d}.npz",
                                frame=np.asarray(pil))
            frame_idx+=1
        poi_ptr+=1

    # 6 | update permanent trail
    overlay_trail[r,c]=PATH_RGB

print(f"first pass: {frame_idx:,} frames → {WORK_DIR}")


# uncomment to clean temp frames
# shutil.rmtree(WORK_DIR)
```

```python

# ───────────────────────── SECOND PASS: NPZ → MP4 ────────────────────────
with imageio.get_writer(VIDEO_OUT, format="FFMPEG", fps=FPS,
                        codec="libx264", quality=10,
                        ffmpeg_log_level="error") as wr:
    for i in range(frame_idx):
        data=np.load(WORK_DIR/f"{i:06d}.npz", mmap_mode="r")
        wr.append_data(data["frame"])
        del data
print("✓ MP4 saved to", VIDEO_OUT)
```
